```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
data = pd.read_csv("dados.csv", sep = ";")
data.shape
```

```python
data.head()
```

```python
plt.hist(data.PIB, bins = 6, color = "gray")
plt.show()
```

```python
srn.histplot(data.PIB, bins=6, kde=True).set(title="PIB")
```

```python
plt.hist(data.VALOREMPENHO, bins=6, color = "gray")
plt.show()
```

```python
srn.histplot(data.VALOREMPENHO, bins=6, kde=True).set(title="Valor Empenho")
```

```python
group = data.sort_values('PIB').head(10)
group = group.iloc[:,[1,2]]
group.plot.bar(x="MUNICIPIO", y="PIB", color = "gray")
```

```python
group = data.sort_values("VALOREMPENHO").head(10)
group = group.iloc[:, [1,3]]
group.plot.bar(x="MUNICIPIO", y="VALOREMPENHO", color="gray")
```
