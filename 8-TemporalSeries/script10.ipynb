```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
dateParse = lambda dates: datetime.strptime(dates, "%Y-%m")
data = pd.read_csv("AirPassengers.csv", parse_dates=["Month"], index_col="Month", date_parser=dateParse)
data.head()
```

```python
ts = data["#Passengers"]
decompose = seasonal_decompose(ts)
```

```python
plt.subplot(3, 1, 1)
plt.plot(decompose.trend, color='blue', linestyle='-')
plt.title("Trend")
plt.xlabel("Time")
plt.ylabel("Value")

plt.subplot(3, 1, 2)
plt.plot(decompose.seasonal, color='green', linestyle='--')
plt.title("Seasonal")
plt.xlabel("Time")
plt.ylabel("Value")

plt.subplot(3, 1, 3)
plt.plot(decompose.resid, color='red', linestyle=':')
plt.title("Residual")
plt.xlabel("Time")
plt.ylabel("Value")

plt.tight_layout()
plt.show()
```
