```python
# Example of Naive Bayes
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv("Credit.csv")
data.head(5)
```

```python
features = data.iloc[:,0:20].values
label = data.iloc[:,20].values
```

```python
columns_to_encode = [0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19]
label_encoders = {}

for col in columns_to_encode:
    le = LabelEncoder()
    features[:, col] = le.fit_transform(features[:, col])
    label_encoders[col] = le
```

```python
featuresTrain, featuresTest, labelTrain, labelTest = train_test_split(features, label, test_size = 0.3, random_state = 0)
featuresTest
```

```python
naiveBayes = GaussianNB()
naiveBayes.fit(featuresTrain, labelTrain)
```

```python
predict = naiveBayes.predict(featuresTest)
```

```python

confusion = confusion_matrix(labelTest, predict)
confusion
```

```python
accuracy = accuracy_score(labelTest, predict)
accuracy
```

```python
newCredit = newData.iloc[:,0:20].values

for col in columns_to_encode:
    newCredit[:, col] = label_encoders[col].fit_transform(newCredit[:, col])

naiveBayes.predict(newCredit)
```
