```python
# C-means
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import numpy as np
import skfuzzy
```

```python
dataset = datasets.load_iris()
```

```python
r = skfuzzy.cmeans(data=dataset.data.T,
                   c = 3,
                   m = 2,
                   error=0.005,
                   maxiter = 1000,
                   init = None)
```

```python
for x in range(150):
    print(r[1][0][x], r[1][1][x], r[1][2][x])
```

```python
predict = r[1].argmax(axis = 0)
confusion = confusion_matrix(dataset.target,predict)
confusion
```
