```python
#K-Medoids
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster import cluster_visualizer
```

```python
dataset = datasets.load_iris()
```

```python
cluster = kmedoids(dataset.data[:, 0:2], [3, 12, 20])
cluster.get_medoids()
```

```python
cluster.process()
group = cluster.get_clusters()
medoids = cluster.get_medoids()
```

```python
s = cluster_visualizer()
s.append_clusters(group, dataset.data[:,0:2])
s.append_cluster(medoids,
                 data = dataset.data[:, 0:2],
                 marker = "*",
                 markersize = 20)
s.show()
```

```python
list_group = []
list_real = []
for i in range(len(group)):
     for j in range(len(group[i])):
        list_group.append(i)
        list_real.append(dataset.target[group[i][j]])
```

```python
list_group = np.asarray(list_group)
list_real = np.asarray(list_real)
confusion = confusion_matrix(list_real, list_group)
confusion
```
